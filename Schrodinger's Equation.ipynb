# Infinite potential well

```python
import numpy as np
from scipy.linalg import eigh
import matplotlib.pyplot as plt

W = 10 # width of well
n = 1000 # number of points
hbar = 1
m = 1
x = np.linspace(0, W, num=n+2)
h = x[1] - x[0] # step size

# Construction of the hamiltonian matrix
H = np.zeros([n, n])
for i in range(n):
    H[i, i] = -2
    if(i-1>=0):
        H[i, i-1] = 1
    if(i+1<n):
        H[i, i+1] = 1
H = -1*(hbar**2)/(2*m*(h**2))*H

# Solution
E, V = eigh(H, eigvals=(0, 10))

# Plots
eigen_funct = np.zeros((n+2))

num_plots = 3
plt.figure(figsize=(8,5))
for i in range(num_plots):
    eigen_funct[1:-1] = V[:, i]
    plt.plot(x, np.abs(eigen_funct)**2)
plt.legend([f'n = {i}' for i in range(num_plots)],loc='upper right')   
plt.title('Solutions to the infinite potential well problem')
plt.ylabel(r'${|\psi^2|}$')
plt.xlabel('x')
plt.show()

```
